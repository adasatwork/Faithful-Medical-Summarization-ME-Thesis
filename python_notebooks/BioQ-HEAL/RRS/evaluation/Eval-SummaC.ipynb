```python
import nltk; nltk.download('punkt_tab')
from summac.model_summac import SummaCZS
model_zs = SummaCZS(granularity="document", model_name="vitc", device="cuda")
```

```python
import pandas as pd

test_df = pd.read_json("data/test-gen3-qwen.jsonl", lines=True)
test_df.info()
```

```python
import numpy as np

def compute_summac(sources, preds):
    scores_zs = model_zs.score(sources, preds)
    return np.mean(scores_zs["scores"])
```

```python
# @title BioBART + Qwen

# score_biobart = compute_summac(
#     test_df["findings"].to_list(),
#     test_df["biobart_gen"].to_list(),
# )

score_qwen = compute_summac(
    test_df["findings"].to_list(),
    test_df["qwen_gen"].to_list(),
)

# score_qwnx = compute_summac(
#     test_df["findings"].to_list(),
#     test_df["qwen3x_gen"].to_list(),
# )
```

```python
# print(f"SummaC Score for BioBART  : {score_biobart:.4f}")
print(f"SummaC Score for Qwen3-4B : {score_qwen:.4f}")
# print(f"SummaC Score for BioQ-HEAL: {score_qwnx:.4f}")
```

```python
# @title Pegasus + Qwen

score_pegasus = compute_summac(
    test_df["findings"].to_list(),
    test_df["pegasus_gen"].to_list(),
)

score_qwen = compute_summac(
    test_df["findings"].to_list(),
    test_df["qwen3_gen"].to_list(),
)

score_qwnx = compute_summac(
    test_df["findings"].to_list(),
    test_df["qwen3x_gen"].to_list(),
)
```
